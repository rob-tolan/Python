```python
#To solve a system with higher-order derivatives, you will first write a cascading system of simple first-order equations then use them in your differential function.
#e.g d*3ydt*3 = C
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
def f(t, y, c):
    dydt = [y[1], y[2], c[0]]
    return dydt
tspan = np.linspace(0, 8, 50)
yinit = [6, 2, -4]
c = [1.3]
sol = solve_ivp(lambda t, y: f(t, y, c), 
                [tspan[0], tspan[-1]], yinit, t_eval=tspan)
print(sol)
```
